```python
import os
import json
import numpy as np

import matplotlib.pyplot as plt
```

```python
with open("experiment_conf_aime2024_Llama-3.2-1B-Instruct_new.json", "r", encoding="utf-8") as f:
    aime24llama = json.load(f)
with open("experiment_conf_math500_Llama-3.2-1B-Instruct.json", "r", encoding="utf-8") as f:
    math500llama = json.load(f)
with open("experiment_conf_aime2024_Qwen2.5-Math-7B-Instruct.json", "r", encoding="utf-8") as f:
    aime24qwen = json.load(f)
with open("experiment_conf_math500_Qwen2.5-Math-7B-Instruct_new.json", "r", encoding="utf-8") as f:
    math500qwen = json.load(f)
```

```python
conf_outs1, conf_outs2, conf_outs3, conf_outs4  = [o['confidence'] for o in aime24llama], [o['confidence'] for o in math500llama], [o['confidence'] for o in aime24qwen], [o['confidence'] for o in math500qwen]

tok_entropy_per_pos1 = [x['token_entropy_per_pos'] for x in conf_outs1]
tok_entropy_per_pos2 = [x['token_entropy_per_pos'] for x in conf_outs2]
tok_entropy_per_pos3 = [x['token_entropy_per_pos'] for x in conf_outs3]
tok_entropy_per_pos4 = [x['token_entropy_per_pos'] for x in conf_outs4]


tok_conf_per_pos1 = [x['token_confidence_per_pos'] for x in conf_outs1]
tok_conf_per_pos2 = [x['token_confidence_per_pos'] for x in conf_outs2]
tok_conf_per_pos3 = [x['token_confidence_per_pos'] for x in conf_outs3]
tok_conf_per_pos4 = [x['token_confidence_per_pos'] for x in conf_outs4]

atc1 = [x['avg_trace_confidence'] for x in conf_outs1]
atc2 = [x['avg_trace_confidence'] for x in conf_outs2]
atc3 = [x['avg_trace_confidence'] for x in conf_outs3]
atc4 = [x['avg_trace_confidence'] for x in conf_outs4]

lowest_group_conf1 = [x['reasoning_lowest_group_confidence'] for x in conf_outs1] 
lowest_group_conf2 = [x['reasoning_lowest_group_confidence'] for x in conf_outs2] 
lowest_group_conf3 = [x['reasoning_lowest_group_confidence'] for x in conf_outs3] 
lowest_group_conf4 = [x['reasoning_lowest_group_confidence'] for x in conf_outs4] 

tail_conf1 = [x['reasoning_tail_confidence'] for x in conf_outs1] 
tail_conf2 = [x['reasoning_tail_confidence'] for x in conf_outs2] 
tail_conf3 = [x['reasoning_tail_confidence'] for x in conf_outs3] 
tail_conf4 = [x['reasoning_tail_confidence'] for x in conf_outs4] 

success_probs1 = [o['success_prob'] for o in aime24llama]
success_probs2 = [o['success_prob'] for o in math500llama]
success_probs3 = [o['success_prob'] for o in aime24qwen]
success_probs4 = [o['success_prob'] for o in math500qwen]

top_group_conf1 = [x['reasoning_highest_group_confidene'] for x in conf_outs1]
top_group_conf4 = [x['reasoning_highest_group_confidene'] for x in conf_outs4]


print(len(atc1))
print(len(lowest_group_conf1))
print(len(tail_conf1))
print(len(success_probs1))
print(len(top_group_conf1))

#reasoning_prefix = [o['reasoning_prefix'] for o in output_data]

#print(len(prm_scores))
#print(len(avg_trace_confidence))
#print(len(token_confidence_per_pos))
#print(len(avg_trace_confidence))

#print(prm_scores[0:5])
#print(avg_trace_confidence[0:5])
```
